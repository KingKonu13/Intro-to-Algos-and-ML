``` check_if_symmetric(string) ```
Return ```True``` if the input string is symmetric (i.e. a palindrome),
and False if not. You can ignore capitalization.

````python
def check_if_sym(word):
    if not isinstance(word, str):
        raise TypeError("must be a str")
    return word == word[::-1]
````

````python
def test_check_if_sym():
    assert check_if_sym("lol") == True
    assert check_if_sym("l o l") == True
    assert check_if_sym(1) == "not a string"
    assert check_if_sym("fish") == False
    assert check_if_sym("") == True
    assert check_if_sym('!ab123 4 321ba!') == True
    print("All tests passed")
````

``` convert_to_numbers(string) ```
Return an array of numbers corresponding to letters in a
string where space = 0, a = 1, b = 2, and so on. For
example, ```convert_to_numbers('a cat')``` should return
```[1,0,3,1,20]```

````python
def convert_to_numbers(word):
    if not isinstance(word, str):
        raise TypeError("please provide a valid input")

    word = word.lower()
    value_array = []
    
    mapping_dict = {
  ' ': 0,'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5,
  'f': 6, 'g': 7, 'h': 8, 'i': 9, 'j': 10,
  'k': 11, 'l': 12, 'm': 13, 'n': 14, 'o': 15,
  'p': 16, 'q': 17, 'r': 18, 's': 19, 't': 20,
  'u': 21, 'v': 22, 'w': 23, 'x': 24, 'y': 25,
  'z': 26
  }
   
    for char in word:
        if char in mapping_dict:
            value_array.append(mapping_dict[char])

    return value_array
````

````python
def test_convert_to_numbers():
    assert convert_to_numbers("a cat") == [1,0,3,1,20]
    assert convert_to_numbers(1) == "please provide a valid input"
    assert convert_to_numbers(" ") == [0]
    assert convert_to_numbers("") == []
    print("All tests passed")
````

````python
test_convert_to_numbers()
````

```get_intersection(array1, array2)```
Return an array consisting of the elements that are in both
```array1``` and ```array2```. There should not be any repeated
elements in the output array

````python
def get_intersection(arr1, arr2):
    intersection_arr = []
    for o in arr1:
        if o in intersection_arr:
            continue
        else:
            if o in arr2:
                intersection_arr.append(o)
    return intersection_arr
````

````python
def test_get_intersection():
    assert get_intersection([1, 2, 3, 4], [3, 4, 5, 6]) == [3, 4]
    assert get_intersection([1, 2, 3], [4, 5, 6]) == []
    assert get_intersection([], [1, 2, 3]) == []
    assert get_intersection([1, 2, 3], []) == []
    assert get_intersection([], []) == []
    assert sorted(get_intersection([1, 2, 2, 3, 4], [2, 2, 3, 5])) == [2, 3]
    assert sorted(get_intersection(["a", "b", "c"], ["b", "c", "d"])) == ["b", "c"]
    print("All tests passed!")
````

```count_characters(string)```
Count the number of each character in a string and
return the counts in a dictionary. Lowercase and
uppercase letters should not be treated differently. For
example, ```count_characters('A cat!!!')``` should
return the ```dictionary { 'a': 2, 'c': 1, 't': 1,
' ': 1, '!': 3 }``` .

````python
def count_characters(string): 
    char_count_dict = {}
    lwstr = [o.lower() for o in string]
    for o in lwstr:
        if o in char_count_dict:
            char_count_dict[o] += 1
        else:
            char_count_dict[o] = 1
    return char_count_dict
````

````python
def test_count_characters():
    # Test basic functionality with mixed case
    assert count_characters('A cat!!!') == {'a': 2, 'c': 1, 't': 1, ' ': 1, '!': 3}
    
    # Test empty string
    assert count_characters('') == {}
    
    # Test string with only one character
    assert count_characters('aaa') == {'a': 3}
    
    # Test with numbers and special characters
    assert count_characters('Hello, World! 123') == {
        'h': 1, 'e': 1, 'l': 3, 'o': 2, ',': 1, 
        ' ': 2, 'w': 1, 'r': 1, 'd': 1, '!': 1,
        '1': 1, '2': 1, '3': 1
    }
    
    # Test with uppercase and lowercase mixed
    assert count_characters('AbCdEf') == {'a': 1, 'b': 1, 'c': 1, 'd': 1, 'e': 1, 'f': 1}
    
    # Test with non-ASCII characters
    assert count_characters('Café') == {'c': 1, 'a': 1, 'f': 1, 'é': 1}
    
    # Test with whitespace characters
    assert count_characters('  \t\n') == {' ': 2, '\t': 1, '\n': 1}
    
    print("All tests passed!")
````

```is_prime(N)```
Check whether an input ```integer N > 1``` is prime by checking
whether there exists some ```n ∈ {2, 3, 4, . . . , ⌊N/2⌋}``` such that
```n divides N.``` (The ⌊ ⌋ symbol denotes the floor function.)

````python
def is_prime(N):
    if N <= 1:
        raise ValueError("input must be an integer larger than one")
    factors = 0
    for o in range(2, (N//2) + 1):
        if N % o == 0:
            return False  
    return True
````

````python
def test_is_prime():
    # Test small prime numbers
    assert is_prime(2) == True
    assert is_prime(3) == True
    assert is_prime(5) == True
    assert is_prime(7) == True
    assert is_prime(11) == True
    assert is_prime(13) == True
    assert is_prime(17) == True
    assert is_prime(19) == True
    
    # Test small non-prime numbers
    assert is_prime(4) == False
    assert is_prime(6) == False
    assert is_prime(8) == False
    assert is_prime(9) == False
    assert is_prime(10) == False
    assert is_prime(12) == False
    assert is_prime(15) == False
    
    # Test larger prime numbers
    assert is_prime(97) == True
    assert is_prime(101) == True
    
    # Test larger non-prime numbers
    assert is_prime(100) == False
    assert is_prime(999) == False
    
    # Test edge cases
    try:
        is_prime(1)  # 1 is not a prime by definition
        assert False, "Should raise an error for N=1"
    except ValueError:
        pass
        
    try:
        is_prime(0)  # 0 is not a prime
        assert False, "Should raise an error for N=0"
    except ValueError:
        pass
        
    try:
        is_prime(-5)  # Negative numbers aren't prime
        assert False, "Should raise an error for negative N"
    except ValueError:
        pass
    
    print("All tests passed!")
````
